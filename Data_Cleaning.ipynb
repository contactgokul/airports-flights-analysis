```python
import pandas as pd
import json    
```

```python
# Mapping the json file
json_data_file = "static/data/airlines_data_json_file.json"

# reading and load the JSON file into a data frame
airline_json_df = pd.read_json(json_data_file, orient='columns')

# View the first five rows
airline_json_df.head(5)
```

```python
#printing the column/values for the required fields
print(airline_json_df["carrier"][0]["name"])
print(airline_json_df["carrier"][0]["code"])
print(airline_json_df["time"][0]["label"])
print(airline_json_df["time"][0]["year"])
print(airline_json_df["time"][0]["month"])
print(airline_json_df["statistics"][0]["flights"]["total"])
```

```python
#reading the created json df for specific columns required and create/store in a new dataframe
airline_carrier_details_df = pd.read_json(airline_json_df["carrier"].to_json(), orient="index")
airline_carrier_details_df.head(5)
```

```python
airline_stats_df = pd.read_json(airline_json_df["statistics"].to_json(), orient="index")
airline_stats_df.head(5)
```

```python
airline_flight_stats_df = pd.read_json(airline_stats_df["flights"].to_json(), orient="index")
airline_flight_stats_df.head(5)
```

```python
airline_time_df = pd.read_json(airline_json_df["time"].to_json(), orient="index")
airline_time_df.head(5)
```

```python
airline_clean_data_df = airline_carrier_details_df[["name", "code"]].copy()
airline_clean_data_df["year_month"] = airline_time_df[["label"]].copy()
airline_clean_data_df["total_flights"] = airline_flight_stats_df[["total"]].copy()
airline_clean_data_df.head()
```

```python
# Mapping has changed the datatypes of the columns to strings
airline_clean_data_df.dtypes
```

```python
airline_clean_data_df.to_csv("static/data/airline_data_csv_file.csv", index=False)
```
